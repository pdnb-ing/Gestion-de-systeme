```python
import random
import string

def generer_mdp_simple(longueur=8):
    # On combine les lettres minuscules (a-z) et majuscules (A-Z)
    caracteres = string.ascii_letters

    # On choisit aléatoirement 'longueur' caractères dans cette liste
    mot_de_passe = ''.join(random.choice(caracteres) for _ in range(longueur))

    return mot_de_passe

# Test de la fonction
print(generer_mdp_simple())
```

```python
import secrets
import string

def generer_mdp_chiffres(longueur=10):
    # Définition des jeux de caractères
    lettres = string.ascii_letters
    chiffres = string.digits
    tous_caracteres = lettres + chiffres

    # 1. On assure la présence de 2 chiffres minimum
    mdp = [secrets.choice(chiffres) for _ in range(2)]

    # 2. On complète avec des caractères aléatoires (lettres ou chiffres)
    # jusqu'à atteindre la longueur souhaitée
    mdp += [secrets.choice(tous_caracteres) for _ in range(longueur - 2)]

    # 3. On mélange la liste pour que les chiffres ne soient pas toujours au début
    secrets.SystemRandom().shuffle(mdp)

    return ''.join(mdp)

# Test de la fonction
print(generer_mdp_chiffres(10))
```

```python
import string
import secrets
from collections import Counter

class PasswordAnalyzer:
    def __init__(self):
        self.alphabet = string.ascii_letters + string.digits + "!@#$%&*"
        self.passwords = []

    def generate_batch(self, count=1000, length=12):
        """Génère un lot de mots de passe forts basé sur votre logique"""
        self.passwords = []
        for _ in range(count):
            while True:
                pwd = ''.join(secrets.choice(self.alphabet) for _ in range(length))
                # Validation : au moins un chiffre et un caractère spécial
                if (any(c.isdigit() for c in pwd) and any(c in "!@#$%&*" for c in pwd)):
                    self.passwords.append(pwd)
                    break
        return self.passwords

    def analyze(self):
        """Analyse les statistiques des mots de passe générés"""
        if not self.passwords:
            return "Aucun mot de passe à analyser."

        total_passwords = len(self.passwords)
        all_chars = "".join(self.passwords)
        total_chars = len(all_chars)

        # 1. Longueur moyenne
        avg_length = total_chars / total_passwords

        # 2. Fréquence des caractères
        char_freq = Counter(all_chars)

        # 3. Diversité (Unicité et Entropy-check simple)
        unique_pwds = len(set(self.passwords))

        # Répartition par catégories
        cats = {
            "Majuscules": sum(1 for c in all_chars if c.isupper()),
            "Minuscules": sum(1 for c in all_chars if c.islower()),
            "Chiffres": sum(1 for c in all_chars if c.isdigit()),
            "Spéciaux": sum(1 for c in all_chars if c in "!@#$%&*")
        }

        # --- Affichage des résultats ---
        print(f"=== ANALYSE DE {total_passwords} MOTS DE PASSE ===")
        print(f"| Longueur moyenne         : {avg_length:.2f}")
        print(f"| Mots de passe uniques    : {unique_pwds} ({unique_pwds/total_passwords*100:.1f}%)")
        print("-" * 40)

        print("DIVERSITÉ DES CATÉGORIES :")
        for cat, val in cats.items():
            pourcentage = (val / total_chars) * 100
            print(f" - {cat:12}: {val} ({pourcentage:.1f}%)")

        print("-" * 40)
        print("TOP 10 DES CARACTÈRES LES PLUS FRÉQUENTS :")
        for char, count in char_freq.most_common(10):
            print(f" '{char}' : {count} fois")

# --- Exécution ---
if __name__ == "__main__":
    analyzer = PasswordAnalyzer()

    # Génération de 1000 mots de passe de longueurs variées (ex: entre 8 et 16)
    # Pour l'analyse, on peut simuler des longueurs aléatoires
    print("Génération en cours...\n")
    analyzer.generate_batch(count=1000, length=12)

    # Lancement de l'analyse
    analyzer.analyze()
```

```python
import hashlib
import secrets

class Etudiant:
    def __init__(self, id_etudiant, nom, prenom, email, mot_de_passe):
        """Initialise l'étudiant et hache immédiatement son mot de passe"""
        self.id = id_etudiant
        self.nom = nom
        self.prenom = prenom
        self.email = email
        # On ne stocke jamais le mot de passe en clair
        self.mot_de_passe_hash = self._creer_hash(mot_de_passe)

    def _creer_hash(self, mot_de_passe):
        """Méthode interne pour générer un hash avec sel """
        salt = secrets.token_bytes(16)
        pwd_hash = hashlib.pbkdf2_hmac('sha256', mot_de_passe.encode(), salt, 100000)
        # On stocke le sel et le hash ensemble pour la vérification future
        return f"{salt.hex()}:{pwd_hash.hex()}"

    def verifier_mot_de_passe(self, mot_de_passe_tente):
        """Vérifie si le mot de passe tenté correspond au hash stocké"""
        try:
            salt_hex, stored_hash_hex = self.mot_de_passe_hash.split(':')
            salt = bytes.fromhex(salt_hex)
            # On génère un nouveau hash avec le sel extrait et le mot de passe tenté
            nouveau_hash = hashlib.pbkdf2_hmac('sha256', mot_de_passe_tente.encode(), salt, 100000)
            return nouveau_hash.hex() == stored_hash_hex
        except Exception:
            return False

    def changer_mot_de_passe(self, ancien_mdp, nouveau_mdp):
        """Change le mot de passe uniquement si l'ancien est correct"""
        if self.verifier_mot_de_passe(ancien_mdp):
            self.mot_de_passe_hash = self._creer_hash(nouveau_mdp)
            print(f"Succès : Le mot de passe de {self.prenom} a été modifié.")
            return True
        else:
            print("Erreur : L'ancien mot de passe est incorrect.")
            return False

# --- Exemple d'utilisation ---
if __name__ == "__main__":
    # Création d'une instance
    nouvel_etu = Etudiant("ETU001", "Joseph", "Pierre", "p.Joseph@ecole.fr", "MonSuperMDP123!")

    # 1. Test de vérification
    print(f"Vérification (Correct) : {nouvel_etu.verifier_mot_de_passe('MonSuperMDP123!')}")
    print(f"Vérification (Faux)    : {nouvel_etu.verifier_mot_de_passe('mauvaismdp')}")

    # 2. Test de changement
    nouvel_etu.changer_mot_de_passe("MonSuperMDP123!", "NouveauPassword456*")

    # 3. Vérification du nouveau mot de passe
    print(f"Vérification nouveau : {nouvel_etu.verifier_mot_de_passe('NouveauPassword456*')}")
```

```python
import csv

class GestionnaireEtudiants:
    def __init__(self):
        """Initialise une liste vide pour stocker les objets Etudiant"""
        self.etudiants = []

    def ajouter_etudiant(self, etudiant):
        """Ajoute un objet Etudiant à la liste"""
        if isinstance(etudiant, Etudiant):
            self.etudiants.append(etudiant)
        else:
            print("Erreur : L'objet ajouté doit être une instance de la classe Etudiant")

    def sauvegarder_csv(self, nom_fichier):
        """Sauvegarde les données des étudiants dans un fichier CSV (sans le mot de passe en clair)"""
        fieldnames = ['id', 'nom', 'prenom', 'email', 'mot_de_passe_hash']

        try:
            with open(nom_fichier, 'w', newline='', encoding='utf-8') as csvfile:
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                writer.writeheader()
                for etu in self.etudiants:
                    writer.writerow({
                        'id': etu.id,
                        'nom': etu.nom,
                        'prenom': etu.prenom,
                        'email': etu.email,
                        'mot_de_passe_hash': etu.mot_de_passe_hash
                    })
            print(f"Sauvegarde réussie dans {nom_fichier}")
        except Exception as e:
            print(f"Erreur lors de la sauvegarde : {e}")

    def charger_csv(self, nom_fichier):
        """Charge les étudiants depuis un fichier CSV et les transforme en objets Etudiant"""
        self.etudiants = [] # On vide la liste actuelle
        try:
            with open(nom_fichier, 'r', encoding='utf-8') as csvfile:
                reader = csv.DictReader(csvfile)
                for row in reader:
                    # On crée l'objet Etudiant
                    # Note : Ici, on réinjecte le hash directement.
                    # J'ajoute une petite adaptation pour que le constructeur accepte un hash existant.
                    nouvel_etu = Etudiant(
                        row['id'],
                        row['nom'],
                        row['prenom'],
                        row['email'],
                        "placeholder" # Mot de passe temporaire
                    )
                    # On remplace le hash généré par celui chargé depuis le CSV
                    nouvel_etu.mot_de_passe_hash = row['mot_de_passe_hash']
                    self.etudiants.append(nouvel_etu)
            print(f"{len(self.etudiants)} étudiants chargés avec succès.")
        except FileNotFoundError:
            print("Erreur : Fichier non trouvé.")
        except Exception as e:
            print(f"Erreur lors du chargement : {e}")

    def trouver_etudiant_par_email(self, email):
        """Recherche un étudiant dans la liste via son email"""
        for etu in self.etudiants:
            if etu.email.lower() == email.lower():
                return etu
        return None

# --- Exemple de mise en pratique combinée ---

if __name__ == "__main__":
    # 1. Création du gestionnaire
    gestionnaire = GestionnaireEtudiants()

    # 2. Ajout de quelques étudiants (en utilisant la classe Etudiant précédente)
    e1 = Etudiant("ETU00001", "SAINT-LOUIS", "Ericka", "ericka@ius.education", "dplice123")
    e2 = Etudiant("ETU00002", "Bernard", "Bruno", "bruno@ius.education", "BrunoPassword456!")

    gestionnaire.ajouter_etudiant(e1)
    gestionnaire.ajouter_etudiant(e2)

    # 3. Sauvegarde
    gestionnaire.sauvegarder_csv("base_donnees_etudiants.csv")

    # 4. Simulation de rechargement
    nouveau_gestionnaire = GestionnaireEtudiants()
    nouveau_gestionnaire.charger_csv("base_donnees_etudiants.csv")

    # 5. Test de recherche et de vérification
    print("\n--- Test du système ---")
    cible = nouveau_gestionnaire.trouver_etudiant_par_email("ericka@ius.education")
    if cible:
        print(f"Étudiant trouvé : {cible.prenom} {cible.nom}")
        test_mdp = "MdpAlice123!"
        if cible.verifier_mot_de_passe(test_mdp):
            print("Authentification réussie !")
        else:
            print("Échec de l'authentification.")
```

```python
import string

def evaluer_force_mot_de_passe(mot_de_passe):
    score = 0
    points_par_critere = 20 # 5 critères majeurs pour arriver à 100

    # 1. Longueur (Minimum 12 caractères)
    if len(mot_de_passe) >= 12:
        score += points_par_critere
    elif len(mot_de_passe) >= 8:
        score += 10 # Score partiel

    # 2. Diversité des caractères
    has_upper = any(c.isupper() for c in mot_de_passe)
    has_lower = any(c.islower() for c in mot_de_passe)
    has_digit = any(c.isdigit() for c in mot_de_passe)
    has_special = any(c in string.punctuation or c in "!@#$%&*" for c in mot_de_passe)

    if all([has_upper, has_lower, has_digit, has_special]):
        score += points_par_critere
    elif sum([has_upper, has_lower, has_digit, has_special]) >= 3:
        score += 10

    # 3. Absence de séquences simples (abc, 123, etc.)
    sequences = ["123", "abc", "qwerty", "azerty", "012"]
    if not any(seq in mot_de_passe.lower() for seq in sequences):
        score += points_par_critere

    # 4. Absence de mots du dictionnaire (Simulation)
    # En production, on utiliserait une vraie liste de mots
    dictionnaire = ["password", "motdepasse", "admin", "etudiant", "bienvenue"]
    if not any(mot in mot_de_passe.lower() for mot in dictionnaire):
        score += points_par_critere

    # 5. Entropie / Complexité (Pas de répétitions abusives)
    if len(set(mot_de_passe)) / len(mot_de_passe) > 0.7:
        score += points_par_critere

    return min(score, 100)

# --- Test du système ---
tests = [
    "123456",              # Très faible
    "Admin123!",           # Moyen (contient dictionnaire + séquence)
    "Kj9$mP2vL#qR"         # Très fort (12 caractères, diversifié, aléatoire)
]

for mdp in tests:
    print(f"Mot de passe: {mdp:15} | Score: {evaluer_force_mot_de_passe(mdp)}/100")
```

```python
import sys
import string
import secrets
from datetime import datetime

# --- Classes précédentes intégrées ---
# (On suppose ici que les classes Etudiant, GestionnaireEtudiants
# et PasswordManager sont définies au-dessus)

class PasswordManager:
    def __init__(self, master_key):
        self.master_key = master_key # Not directly used for generation, but kept for context

    def generate_password(self, length=12):
        """Génère un mot de passe fort avec lettres, chiffres et symboles."""
        alphabet = string.ascii_letters + string.digits + string.punctuation
        password = ''.join(secrets.choice(alphabet) for i in range(length))
        return password

class ConsoleInterface:
    def __init__(self):
        self.gestionnaire = GestionnaireEtudiants()
        self.pm = None
        self.master_key_verified = False

    def verifier_master_password(self):
        """Sécurité d'entrée : On force l'usage d'un mot de passe maître robuste"""
        print("--- SYSTÈME DE GESTION SÉCURISÉ ---")
        pwd = input("Entrez le mot de passe maître pour déverrouiller le système : ")
        if len(pwd) < 8:
            print("Erreur : Le mot de passe maître est trop court. Accès refusé.")
            sys.exit()

        self.pm = PasswordManager(pwd)
        self.master_key_verified = True
        print("Système déverrouillé avec succès.\n")

    def menu(self):
        while True:
            print("\n" + "="*30)
            print("      MENU PRINCIPAL")
            print("="*30)
            print("1. Générer un mot de passe fort")
            print("2. Ajouter un étudiant (Sécurisé)")
            print("3. Lister les étudiants")
            print("4. Sauvegarder les données (CSV)")
            print("5. Charger les données (CSV)")
            print("6. Audit de sécurité complet")
            print("0. Quitter")

            choix = input("\nChoisissez une option : ")

            if choix == "1":
                self.ui_generer_mdp()
            elif choix == "2":
                self.ui_ajouter_etudiant()
            elif choix == "3":
                self.ui_lister_etudiants()
            elif choix == "4":
                self.gestionnaire.sauvegarder_csv("etudiants_secure.csv")
            elif choix == "5":
                self.gestionnaire.charger_csv("etudiants_secure.csv")
            elif choix == "6":
                self.ui_audit_securite()
            elif choix == "0":
                print("Fermeture sécurisée du système...")
                break
            else:
                print("Option invalide.")

    def ui_generer_mdp(self):
        while True:
            longueur_input = input("Longueur du mot de passe (défaut 12) : ")
            if not longueur_input:
                longueur = 12
                break
            try:
                longueur = int(longueur_input)
                if longueur <= 0:
                    print("La longueur doit être un nombre entier positif.")
                else:
                    break
            except ValueError:
                print("Entrée invalide. Veuillez entrer un nombre entier.")

        mdp = self.pm.generate_password(longueur)
        score = evaluer_force_mot_de_passe(mdp)
        print(f"\nMot de passe généré : {mdp}")
        print(f"Force évaluée : {score}/100")

    def ui_ajouter_etudiant(self):
        print("\n--- Nouvel Étudiant ---")
        id_etu = input("ID : ")
        nom = input("Nom : ")
        prenom = input("Prénom : ")
        email = input("Email : ")

        while True:
            mdp = input("Entrez un mot de passe (ou 'GEN' pour générer) : ")
            if mdp.upper() == "GEN":
                mdp = self.pm.generate_password(14)
                print(f"Mot de passe généré : {mdp}")

            score = evaluer_force_mot_de_passe(mdp)
            if score < 70:
                print(f"Erreur : Mot de passe trop faible ({score}/100). Réessayez.")
            else:
                break

        nouvel_etu = Etudiant(id_etu, nom, prenom, email, mdp)
        self.gestionnaire.ajouter_etudiant(nouvel_etu)
        print("Étudiant ajouté avec succès.")

    def ui_lister_etudiants(self):
        print(f"\n{'ID':<10} | {'Nom':<15} | {'Email':<25}")
        print("-" * 55)
        for etu in self.gestionnaire.etudiants:
            print(f"{etu.id:<10} | {etu.nom:<15} | {etu.email:<25}")

    def ui_audit_securite(self):
        """Analyse la vulnérabilité de la base actuelle"""
        print("\n--- RAPPORT D'AUDIT DE SÉCURITÉ ---")
        nb = len(self.gestionnaire.etudiants)
        if nb == 0:
            print("Aucune donnée à analyser.")
            return

        print(f"Nombre d'utilisateurs : {nb}")
        print(f"Algorithme de hachage : PBKDF2-HMAC-SHA256")
        print(f"Itérations : 100,000")

        # Simulation d'analyse de fuite de sel
        salts = [etu.mot_de_passe_hash.split(':')[0] for etu in self.gestionnaire.etudiants]
        doublons_sel = len(salts) - len(set(salts))

        print(f"Doublons de sel détectés : {doublons_sel}")
        if doublons_sel == 0:
            print("Score de santé : EXCELLENT (Tous les sels sont uniques)")
        else:
            print("Score de santé : CRITIQUE (Réutilisation de sel détectée)")

# --- Lancement ---
if __name__ == "__main__":
    app = ConsoleInterface()
    app.verifier_master_password()
    app.menu()
```

```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from datetime import datetime

def generer_pdf_etudiants(etudiants, nom_fichier="coupons_etudiants.pdf"):
    """Génère un PDF contenant les identifiants et mots de passe en clair pour distribution"""

    c = canvas.Canvas(nom_fichier, pagesize=A4)
    largeur, hauteur = A4
    y = hauteur - 50  # Position de départ verticale

    # --- En-tête du document ---
    c.setFont("Helvetica-Bold", 16)
    c.drawString(50, y, "Coupons d'accès au portail étudiant")
    y -= 20
    c.setFont("Helvetica-Oblique", 10)
    c.drawString(50, y, f"Généré le : {datetime.now().strftime('%d/%m/%Y %H:%M')}")
    y -= 40

    # --- Génération des coupons ---
    c.setFont("Helvetica", 12)

    for etu in etudiants:
        # Si on arrive en bas de page, on crée une nouvelle page
        if y < 150:
            c.showPage()
            y = hauteur - 50
            c.setFont("Helvetica", 12)

        # Dessin d'un cadre pour le coupon
        c.setStrokeColor(colors.black)
        c.rect(40, y - 100, largeur - 80, 110, stroke=1, fill=0)

        # Contenu du coupon
        c.setFont("Helvetica-Bold", 12)
        c.drawString(60, y - 20, f"Étudiant : {etu['prenom']} {etu['nom']}")

        c.setFont("Helvetica", 11)
        c.drawString(60, y - 40, f"Identifiant (ID) : {etu['id']}")
        c.drawString(60, y - 55, f"Email : {etu['email']}")

        c.setDash(3, 3) # Ligne pointillée pour le mot de passe
        c.line(60, y - 65, largeur - 60, y - 65)
        c.setDash(1, 0) # Retour à ligne pleine

        c.setFont("Helvetica-Bold", 11)
        c.setFillColor(colors.red)
        # On utilise 'mot_de_passe_clair' car c'est ce que l'étudiant doit saisir
        c.drawString(60, y - 85, f"MOT DE PASSE TEMPORAIRE : {etu['mot_de_passe_clair']}")

        c.setFillColor(colors.black)
        c.setFont("Helvetica-Oblique", 8)
        c.drawString(60, y - 95, "Note : Changez ce mot de passe dès votre première connexion.")

        y -= 130 # Espacement entre deux coupons

    c.save()
    print(f"\n[Succès] Le fichier PDF '{nom_fichier}' a été généré.")

# --- Intégration dans votre fonction main() ---
# Dans votre boucle main, après avoir généré la liste 'students' :
# generer_pdf_etudiants(students)
```
